# Week 11 — Subobject Classifier (Sets)
We illustrate the subobject classifier in **Sets** using a simple subset.


```python
# Universe and subobject
U = {0,1,2,3,4,5}
A = {1,2,5}

def chi_A(x):
    return 1 if x in A else 0

truth = {0:'false', 1:'true'}

print('Universe U:', U)
print('Subobject A:', A)
print('Characteristic map χ_A: U -> Ω')
for x in sorted(U):
    print(x, '->', truth[chi_A(x)])

# Pullback idea: membership in A is equivalent to χ_A(x)=true
A_from_chi = {x for x in U if chi_A(x) == 1}
print('Recovered A from χ_A:', A_from_chi)
```
